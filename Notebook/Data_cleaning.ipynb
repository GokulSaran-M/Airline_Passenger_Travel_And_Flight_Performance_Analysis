```python
df.reset_index(drop=True)
df['flight_id']= df.index + 1
```

```python
age_range = [0,5,12,18,35,50,100]
age_categ = ['Baby','Child','Teen','Adult','Middle_aged','Senior_citizen']
df['Age_Category'] = pd.cut(df['Age'], bins = age_range, labels = age_categ)
print('done')
```

```python
df[['Age','Age_Category']]
```

```python
df['Passenger_Full_Name'] = df['First Name'] + ' ' + df['Last Name']
df[['First Name', 'Last Name', 'Passenger_Full_Name']]
```

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'])
df['Departure_Year'] = df['Departure Date'].dt.year
df['Departure_Month'] = df['Departure Date'].dt.month
print('done')
```

```python
df[['Departure Date','Departure_Year','Departure_Month']]
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns
```
